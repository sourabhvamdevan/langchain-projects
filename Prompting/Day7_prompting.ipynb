```python
from google import genai
from google.colab import userdata
api_key = userdata.get('GOOGLE_API_KEY')
client = genai.Client(api_key=api_key)

prompt = """
Classify the sentiment of the following review as Positive, Negative, or Neutral.

Review: "The battery life on this new phone easily lasts two days, but the screen is a bit dimmer than I expected."
Sentiment:
"""
response = client.models.generate_content(
    model='gemini-2.5-flash',
    contents=prompt,
)

print("--- Zero-Shot Prompting ---")
print(response.text)

```

```python
from google import genai
from google.genai import types
from google.colab import userdata

client = genai.Client(api_key=userdata.get('GOOGLE_API_KEY'))


config = types.GenerateContentConfig(
    system_instruction="You are a strict, no-nonsense Cyber Security Auditor. You speak bluntly and prioritize absolute security."
)

prompt = "Our team wants to store user passwords in plain text temporarily in our database just while we test a feature. Is that okay?"

response = client.models.generate_content(
    model='gemini-2.5-flash',
    contents=prompt,
    config=config
)

print("--- Persona-Based Prompting ---")
print(response.text)
```

```python
from google import genai
from google.genai import types
from google.colab import userdata

client = genai.Client(api_key=userdata.get('GOOGLE_API_KEY'))

config = types.GenerateContentConfig(
    system_instruction="You are an expert career coach interviewing a candidate for a Software Engineer role. Ask ONE relevant interview question at a time. Wait for the user's answer before asking the next question."
)


chat = client.chats.create(model="gemini-2.5-flash", config=config)

print("--- Interview-Style Prompting (Live Interactive Loop) ---")
print("Type 'exit' to end the interview.\n")


response = chat.send_message("Hello, I am ready for my mock interview.")
print(f"Interviewer: {response.text}\n")


for _ in range(3):
    user_input = input("Your Answer: ")
    if user_input.lower() == 'exit':
        break

    print()
    response = chat.send_message(user_input)
    print(f"Interviewer: {response.text}\n")
```
