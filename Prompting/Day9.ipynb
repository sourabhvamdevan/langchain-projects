```python
import pandas as pd
import warnings
import matplotlib.pyplot as plt

from sklearn.datasets import fetch_california_housing
from ctgan import CTGAN
from table_evaluator import TableEvaluator

warnings.filterwarnings("ignore")



print("=" * 70)
print("STEP 1 : Loading California Housing Dataset")
print("=" * 70)

housing = fetch_california_housing(as_frame=True)
real_data = housing.frame

print(f"\nDataset Shape : {real_data.shape}")
print("\nColumns:")
print(real_data.columns.tolist())

print("\nFirst 5 Rows")
print(real_data.head())



print("\n" + "=" * 70)
print("STEP 2 : Training CTGAN")
print("=" * 70)

discrete_columns = []

model = CTGAN(
    epochs=5,
    verbose=True
)

model.fit(real_data, discrete_columns)

print("\nTraining Completed Successfully!")



print("\n" + "=" * 70)
print("STEP 3 : Generating Synthetic Dataset")
print("=" * 70)

synthetic_data = model.sample(500)

print(f"Synthetic Dataset Shape : {synthetic_data.shape}")



synthetic_data.to_csv("synthetic_california_housing.csv", index=False)

print("\nSynthetic dataset saved as:")
print("synthetic_california_housing.csv")



pd.set_option("display.max_columns", None)
pd.set_option("display.width", 1200)

print("\n" + "=" * 70)
print("REAL DATA (First 5 Rows)")
print("=" * 70)
print(real_data.head().round(3))

print("\n")

print("=" * 70)
print("SYNTHETIC DATA (First 5 Rows)")
print("=" * 70)
print(synthetic_data.head().round(3))



print("\n" + "=" * 70)
print("STATISTICAL COMPARISON")
print("=" * 70)

comparison = pd.DataFrame({
    "Real Mean": real_data.mean(),
    "Synthetic Mean": synthetic_data.mean(),
    "Real Std": real_data.std(),
    "Synthetic Std": synthetic_data.std()
})

print(comparison.round(3))



print("\n" + "=" * 70)
print("STEP 7 : Evaluating Synthetic Data Quality")
print("=" * 70)

real_sample = real_data.sample(
    n=500,
    random_state=42
)

evaluator = TableEvaluator(
    real_sample,
    synthetic_data
)

print("\nGenerating Distribution Plots...\n")

evaluator.visual_evaluation()

print("\nGenerating Correlation Plots...\n")

evaluator.plot_correlation_difference()

print("\nCalculating Similarity Scores...\n")

results = evaluator.evaluate(
    target_col="MedHouseVal",
    target_type="regr"
)

print("\nEvaluation Completed!")



print("\n" + "=" * 70)
print("COLUMN-WISE MEAN DIFFERENCE")
print("=" * 70)

for column in real_data.columns:

    real_mean = real_data[column].mean()
    fake_mean = synthetic_data[column].mean()

    diff = abs(real_mean - fake_mean)

    print(
        f"{column:15s}"
        f" Real={real_mean:10.3f}"
        f" Synthetic={fake_mean:10.3f}"
        f" Difference={diff:10.3f}"
    )



print("\nGenerating Histogram Comparisons...")

for column in real_data.columns:

    plt.figure(figsize=(7,4))

    plt.hist(
        real_data[column],
        bins=30,
        alpha=0.5,
        label="Real"
    )

    plt.hist(
        synthetic_data[column],
        bins=30,
        alpha=0.5,
        label="Synthetic"
    )

    plt.title(column)
    plt.xlabel(column)
    plt.ylabel("Frequency")
    plt.legend()

    plt.tight_layout()
    plt.show()



print("\nReal Correlation Matrix")
print(real_data.corr().round(2))

print("\nSynthetic Correlation Matrix")
print(synthetic_data.corr().round(2))

print("\n" + "=" * 70)
print("PROCESS COMPLETED SUCCESSFULLY")
print("=" * 70)
```
