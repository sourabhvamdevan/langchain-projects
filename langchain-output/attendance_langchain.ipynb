```python
from langchain.tools import tool

@tool
def attendance_percentage(total_classes: int, attended_classes: int):
    """Calculate attendance percentage."""
    percentage = (attended_classes / total_classes) * 100
    return f"Attendance: {percentage:.2f}%"

@tool
def fee_balance(total_fee: float, paid_fee: float):
    """Calculate pending fee amount."""
    balance = total_fee - paid_fee
    return f"Pending Fee: ₹{balance}"

@tool
def cgpa_calculator(total_grade_points: float, total_credits: int):
    """Calculate CGPA."""
    cgpa = total_grade_points / total_credits
    return f"CGPA: {cgpa:.2f}"

@tool
def age_calculator(birth_year: int, current_year: int):
    """Calculate age."""
    age = current_year - birth_year
    return f"Age: {age} years"
```

```python
print(
    attendance_percentage.invoke(
        {
            "total_classes":120,
            "attended_classes":96
        }
    )
)

print(
    fee_balance.invoke(
        {
            "total_fee":50000,
            "paid_fee":35000
        }
    )
)

print(
    cgpa_calculator.invoke(
        {
            "total_grade_points":78,
            "total_credits":9
        }
    )
)

print(
    age_calculator.invoke(
        {
            "birth_year":2005,
            "current_year":2026
        }
    )
)
```

```python
from langchain.tools import tool
from langchain_google_genai import ChatGoogleGenerativeAI
import os

os.environ["GOOGLE_API_KEY"] = "your_key"

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash"
)

@tool
def attendance(total_classes: int, attended_classes: int):
    """Calculate attendance percentage."""
    return (attended_classes / total_classes) * 100

@tool
def fee(total_fee: float, paid_fee: float):
    """Calculate remaining fee."""
    return total_fee - paid_fee

print(
    attendance.invoke(
        {
            "total_classes":100,
            "attended_classes":85
        }
    )
)

print(
    fee.invoke(
        {
            "total_fee":50000,
            "paid_fee":32000
        }
    )
)

response = llm.invoke(
    "Explain why attendance is important for students."
)

print(response.content)
```
